```python
#Ce code implémente une régression linéaire et une régression polynomiale pour prédire un salaire (Salary) en fonction du niveau d'expérience (Level).
#partie1 ex3
%%R
# Importing the dataset
dataset = read.csv('Position_Salaries.csv')
dataset = dataset[2:3]

# Splitting the dataset into the Training set and Test set
# # install.packages('caTools')
# library(caTools)
# set.seed(123)
# split = sample.split(dataset$Salary, SplitRatio = 2/3)
# training_set = subset(dataset, split == TRUE)
# test_set = subset(dataset, split == FALSE)

# Feature Scaling
# training_set = scale(training_set)
# test_set = scale(test_set)

# Fitting Linear Regression to the dataset
lin_reg = lm(formula = Salary ~ ., data = dataset)

# Régression Polynômiale
dataset$Level2 = dataset$Level^2
dataset$Level3 = dataset$Level^3
dataset$Level4 = dataset$Level^4
poly_reg = lm(formula = Salary ~ ., data = dataset)

# Visualising the Linear Regression results
# install.packages('ggplot2')

library(ggplot2)
ggplot() +
  geom_point(aes(x = dataset$Level, y = dataset$Salary), colour = 'red') +
  geom_line(aes(x = dataset$Level, y = predict(lin_reg, newdata = dataset)), colour = 'blue') +
  ggtitle('Truth or Bluff (Linear Regression)') +
  xlab('Level') +
  ylab('Salary')

# Visualising the Polynomial Regression results
# install.packages('ggplot2')
library(ggplot2)
ggplot() +
  geom_point(aes(x = dataset$Level, y = dataset$Salary), colour = 'red') +
  geom_line(aes(x = dataset$Level, y = predict(poly_reg, newdata = dataset)), colour = 'blue') +
  ggtitle('Truth or Bluff (Polynomial Regression)') +
  xlab('Level') +
  ylab('Salary')

# Visualising the Regression Model results (for higher resolution and smoother curve)
# install.packages('ggplot2')
library(ggplot2)
x_grid = seq(min(dataset$Level), max(dataset$Level), 0.1)
ggplot() +
  geom_point(aes(x = dataset$Level, y = dataset$Salary), colour = 'red') +
  geom_line(aes(x = x_grid, y = predict(poly_reg,
          newdata = data.frame(Level = x_grid,
                               Level2 = x_grid^2,
                               Level3 = x_grid^3,
                               Level4 = x_grid^4))),
            colour = 'blue') +
  ggtitle('Truth or Bluff (Polynomial Regression)') +
  xlab('Level') +
  ylab('Salary')

# Predicting a new result with Linear Regression
predict(lin_reg, data.frame(Level = 6.5))

# Predicting a new result with Polynomial Regression
predict(poly_reg, data.frame(Level = 6.5,
                             Level2 = 6.5^2,
                             Level3 = 6.5^3,
                             Level4 = 6.5^4))

```

# Rapport d'Analyse : Comparaison entre Régression Linéaire et Polynomiale pour la Prédiction de Salaires

## Résumé
Ce rapport présente une analyse comparative entre un modèle de régression linéaire simple et un modèle de régression polynomiale pour prédire les salaires en fonction du niveau d'expérience professionnelle. Les données utilisées proviennent du fichier "Position_Salaries.csv".

## Méthodologie

### Données
Le jeu de données comprend les colonnes "Level" (niveau d'expérience) et "Salary" (salaire). Nous avons utilisé l'ensemble des données pour l'entraînement des modèles, sans division en ensembles d'entraînement et de test.

### Modèles implémentés
1. **Régression linéaire simple** : Modèle standard de la forme Salary = β₀ + β₁×Level
2. **Régression polynomiale** : Modèle incluant des termes polynomiaux jusqu'au 4ème degré, de la forme Salary = β₀ + β₁×Level + β₂×Level² + β₃×Level³ + β₄×Level⁴

### Prétraitement des données
Pour la régression polynomiale, nous avons créé des variables additionnelles (Level², Level³, Level⁴) représentant les puissances du niveau d'expérience.

## Résultats visuels

L'analyse graphique révèle des différences significatives entre les deux modèles :

1. **Régression linéaire** : Le modèle linéaire montre une tendance générale croissante, mais ne capture pas les variations non-linéaires présentes dans les données. Il sous-estime les salaires aux niveaux d'expérience très bas et très élevés, et les surestime pour les niveaux intermédiaires.

2. **Régression polynomiale** : Le modèle polynomial de degré 4 s'ajuste beaucoup mieux aux données, capturant efficacement la relation non-linéaire entre le niveau d'expérience et le salaire. La courbe présente une forme en S qui correspond bien à la réalité économique des progressions de salaire.

## Analyse et interprétation

### Forces et faiblesses des modèles

**Régression linéaire**
- Forces : Simplicité, interprétabilité
- Faiblesses : Incapacité à capturer des relations non-linéaires, précision limitée

**Régression polynomiale**
- Forces : Flexibilité, capacité à modéliser des relations non-linéaires complexes
- Faiblesses : Risque de surapprentissage avec des polynômes d'ordre élevé, extrapolation potentiellement instable hors de la plage des données

### Ajustement aux données
La régression polynomiale offre un ajustement visuellement supérieur aux données, ce qui suggère que la relation entre le niveau d'expérience et le salaire n'est pas linéaire. Ceci est cohérent avec la réalité économique, où les progressions salariales tendent à être plus importantes aux niveaux d'expérience élevés.

## Conclusions

1. La régression polynomiale de degré 4 est nettement supérieure à la régression linéaire pour modéliser la relation entre le niveau d'expérience et le salaire dans ce jeu de données.

2. Les visualisations confirment que la relation sous-jacente est clairement non-linéaire, justifiant l'utilisation d'un modèle polynomial.

3. Pour des prédictions précises, particulièrement aux extrémités de l'échelle d'expérience, le modèle polynomial est recommandé.
