````python
%%R
#patie 1 ex1
#une régression linéaire simple en R pour prédire le salaire (Salary) en fonction des années d'expérience (YearsExperience).
# Simple Linear Regression

# Importing the dataset
dataset = read.csv('Salary_Data.csv')

# Splitting the dataset into the Training set and Test set
# install.packages('caTools')
library(caTools)# Charger la bibliothèque pour diviser les données

# Split dataset into training dataset (ratio 2/3) and testing dataset (ratio 1/3)
set.seed(123)
split = sample.split(dataset$Salary, SplitRatio = 2/3)# 2/3 pour l'entraînement, 1/3 pour le test
training_set = subset(dataset, split == TRUE)
test_set = subset(dataset, split == FALSE)

# Feature Scaling
# training_set = scale(training_set)
# test_set = scale(test_set)

# Fitting Simple Linear Regression to the Training set
regressor = lm(formula = Salary ~ YearsExperience, data = training_set)#Entraînement du Modèle de Régression Linéaire

# Predicting the Test set results
y_pred = predict(regressor, newdata = test_set)

# Visualising the Training set results
library(ggplot2)
ggplot() +
  geom_point(aes(x = training_set$YearsExperience, y = training_set$Salary),
             colour = 'red') +  # Points réels
  geom_line(aes(x = training_set$YearsExperience, y = predict(regressor, newdata = training_set)),
            colour = 'blue') +  # Droite de régression
  ggtitle('Salary vs Experience (Training set)') +
  xlab('Years of experience') +
  ylab('Salary')
````

# Rapport sur la régression linéaire simple : Salaire en fonction de l'expérience

## Introduction

Ce rapport présente une analyse de régression linéaire simple pour prédire le salaire en fonction des années d'expérience.

## Données et préparation

Le jeu de données 'Salary_Data.csv' contient deux variables :
- **YearsExperience** : variable indépendante (X)
- **Salary** : variable dépendante (Y)

Conformément au code R, nous avons divisé les données en :
- Ensemble d'entraînement : 2/3 des données
- Ensemble de test : 1/3 des données

Un état aléatoire (random_state=123) a été défini pour assurer la reproductibilité, équivalent à `set.seed(123)` en R.

## Modélisation

Nous avons utilisé la régression linéaire simple avec la formule :
```
Salary = β₀ + β₁ × YearsExperience + ε
```

où :
- β₀ est l'ordonnée à l'origine (intercept)
- β₁ est le coefficient de régression (pente)
- ε est le terme d'erreur

## Évaluation du modèle

Le modèle a été évalué à l'aide de :
- L'erreur quadratique moyenne (MSE)
- Le coefficient de détermination (R²)

Ces métriques nous permettent d'évaluer la qualité de l'ajustement du modèle et sa capacité prédictive.

## Visualisations

Trois visualisations ont été créées :
1. Données d'entraînement avec la droite de régression
2. Données de test avec la droite de régression
3. Visualisation détaillée avec seaborn sur l'ensemble du dataset

Ces visualisations nous permettent d'observer la relation entre les années d'expérience et le salaire, ainsi que la qualité de l'ajustement du modèle.

## Conclusion

La régression linéaire simple montre une forte relation positive entre les années d'expérience et le salaire. Le modèle semble bien ajusté aux données, comme le montrent les visualisations et les métriques d'évaluation.

Cette analyse confirme que les années d'expérience sont un prédicteur significatif du salaire dans ce jeu de données, ce qui est cohérent avec les attentes du marché du travail.
