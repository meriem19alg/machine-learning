```python
%%R
# Charger les bibliothèques nécessaires
library(rpart)
library(rpart.plot)

# 1. Charger les données iris
data(iris)

# 2. Afficher les premiers enregistrements
head(iris)

# 3. Créer l'arbre de décision
arbre_iris <- rpart(Species ~ ., data = iris, method = "class")

# Visualiser l'arbre
rpart.plot(arbre_iris, main = "Arbre de décision sur les données Iris")

# 4. Visualiser la qualité de la classification
# Afficher les prédictions sur les mêmes données (validation interne)
predictions <- predict(arbre_iris, type = "class")

# Créer une matrice de confusion pour évaluer la qualité de la classification
table(predictions, iris$Species)

# Calculer la précision
accuracy <- sum(predictions == iris$Species) / length(iris$Species)
cat("La précision de la classification est : ", accuracy, "\n")
```

Ce code en R crée un arbre de décision pour classifier les fleurs de l'ensemble de données Iris en fonction de leurs caractéristiques. Il utilise le package rpart pour générer l'arbre et rpart.plot pour le visualiser. Enfin, il évalue la qualité du modèle à l'aide d'une matrice de confusion et calcule la précision de la classification.
