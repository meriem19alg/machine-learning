```python
%%R
# Charger les bibliothèques nécessaires
library(rpart)
library(rpart.plot)

# Créer le jeu de données directement
tennis <- data.frame(
  Ciel = c("Ensoleille", "Ensoleille", "Couvert", "Pluie", "Pluie", "Pluie",
           "Couvert", "Ensoleille", "Ensoleille", "Pluie", "Ensoleille",
           "Couvert", "Couvert", "Pluie"),
  Temperature = c(27.5, 25, 26.5, 20, 19, 17.5, 17, 21, 19.5, 22.5, 22.5,
                  21, 25.5, 20.5),
  Humidite = c(85, 90, 86, 96, 80, 70, 65, 95, 70, 80, 70, 90, 75, 91),
  Vent = c("Faible", "Fort", "Faible", "Faible", "Faible", "Fort", "Fort",
           "Faible", "Faible", "Faible", "Fort", "Fort", "Faible", "Fort"),
  Jouer = c("Non", "Non", "Oui", "Oui", "Oui", "Non", "Oui", "Non", "Oui",
            "Oui", "Oui", "Oui", "Oui", "Non")
)

# Convertir les variables catégorielles en facteurs
tennis$Ciel <- as.factor(tennis$Ciel)
tennis$Vent <- as.factor(tennis$Vent)
tennis$Jouer <- as.factor(tennis$Jouer)  # La variable cible doit être un facteur

# Création des arbres avec différentes valeurs de minsplit
arbre1 <- rpart(Jouer ~ ., data = tennis, control = rpart.control(minsplit = 1))
arbre2 <- rpart(Jouer ~ ., data = tennis, control = rpart.control(minsplit = 2))
arbre5 <- rpart(Jouer ~ ., data = tennis, control = rpart.control(minsplit = 5))
arbre20 <- rpart(Jouer ~ ., data = tennis, control = rpart.control(minsplit = 20))

# Visualisation des arbres
par(mfrow = c(2, 2))  # Pour afficher 4 graphiques en même temps

rpart.plot(arbre1, main = "Arbre avec minsplit = 1")
rpart.plot(arbre2, main = "Arbre avec minsplit = 2")
rpart.plot(arbre5, main = "Arbre avec minsplit = 5")
rpart.plot(arbre20, main = "Arbre avec minsplit = 20")

# Réinitialisation de la disposition des graphiques
par(mfrow = c(1, 1))


```
