```python
%%R
# Installer et charger le package ISLR
install.packages("ISLR")  # À exécuter une seule fois
library(ISLR)

# Charger les données
data(Default)

# Vérifier les premières lignes du dataset
head(Default)
# Afficher un résumé des données
summary(Default)
# Transformer la variable default en 0 (No) et 1 (Yes)
Default$default <- ifelse(Default$default == "Yes", 1, 0)

# Vérifier la transformation
table(Default$default)
# Ajuster un modèle logistique avec balance comme seule variable explicative
logit_model_balance <- glm(default ~ balance, data = Default, family = binomial)

# Résumé du modèle
summary(logit_model_balance)
# Créer un dataframe pour les nouvelles valeurs de balance
new_data <- data.frame(balance = c(1000, 1500, 2000, 3000))

# Prédire la probabilité de défaut
new_data$predicted_prob <- predict(logit_model_balance, newdata = new_data, type = "response")

print(new_data)
# Créer le tableau de contingence
table_contingence <- table(Default$default, Default$student)
print(table_contingence)
# Probabilité conditionnelle par groupe
prop_table <- prop.table(table_contingence, margin = 2)
print(prop_table)
# Ajuster le modèle
logit_model_student_balance <- glm(default ~ student + balance, data = Default, family = binomial)

# Résumé du modèle
summary(logit_model_student_balance)
# Ajuster un modèle logistique avec les 3 variables explicatives
logit_model_full <- glm(default ~ student + balance + income, data = Default, family = binomial)

# Résumé du modèle
summary(logit_model_full)
AIC(logit_model_balance, logit_model_student_balance, logit_model_full)
```

Régression Logistique sur le Dataset Default

Objectif : Prédire le défaut de paiement en fonction de balance, statut étudiant et revenu.

Méthodes :
Ajustement de modèles logistiques avec différentes variables.
Prédiction de la probabilité de défaut.
Comparaison des modèles avec l’AIC.

Résultats :
Balance est le facteur le plus important : plus elle est élevée, plus le risque de défaut augmente.
Le statut d’étudiant a peu d’impact.
Le modèle avec balance et revenu est le plus performant.

Conclusion : La balance influence fortement le risque de défaut, contrairement au statut d’étudiant.
