```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import os

# ---------------------------
# Partie 1 : Classification synthétique
# ---------------------------

try:
    # Charger les données d'entraînement
    data = pd.read_csv("synth_train.txt", sep='\s+')
    data.columns = data.columns.str.strip()  # Enlever les espaces des noms de colonnes

    # Vérifier les valeurs manquantes
    missing_count = data.isnull().sum().sum()
    if missing_count > 0:
        print(f"Attention: {missing_count} valeurs manquantes détectées dans les données d'entraînement")
        # Possibilité d'ajouter un traitement pour les valeurs manquantes ici

    # Vérification et conversion en facteur de la colonne 'y'
    if 'y' in data.columns:
        data['y'] = data['y'].astype('category')
    else:
        raise ValueError("Erreur : La colonne y n'existe pas dans les données d'apprentissage")

    # Charger les données de test
    data_test = pd.read_csv("synth_test.txt", sep='\s+')
    data_test.columns = data_test.columns.str.strip()

    # Vérifier les valeurs manquantes dans les données de test
    missing_count_test = data_test.isnull().sum().sum()
    if missing_count_test > 0:
        print(f"Attention: {missing_count_test} valeurs manquantes détectées dans les données de test")

    # Vérification et conversion en facteur de la colonne 'y'
    if 'y' in data_test.columns:
        data_test['y'] = data_test['y'].astype('category')
    else:
        raise ValueError("Erreur : La colonne y n'existe pas dans les données de test")

except FileNotFoundError as e:
    print(f"Erreur de chargement des fichiers: {e}")
    print("Vérifiez que les fichiers synth_train.txt et synth_test.txt sont présents dans le répertoire de travail.")
    exit(1)
except Exception as e:
    print(f"Erreur lors du chargement des données: {e}")
    exit(1)

# Vérification des dimensions des données
print(f"Dimensions des données d'apprentissage : {data.shape}")
print(f"Dimensions des données de test : {data_test.shape}")

# Étape 2 : Construire l'arbre de classification standard pour synth_train
X = data.drop(columns=['y'])
y = data['y']

# Arbre de décision standard
tree_standard = DecisionTreeClassifier(random_state=42)
tree_standard.fit(X, y)

# Visualiser l'arbre de décision standard
plt.figure(figsize=(20, 10))
# Limiter la profondeur de l'arbre affiché si l'arbre est trop profond
max_depth_to_plot = min(5, tree_standard.tree_.max_depth)
plot_tree(tree_standard, filled=True, feature_names=X.columns,
          class_names=[str(cls) for cls in y.cat.categories],
          rounded=True, max_depth=max_depth_to_plot)
plt.title("Arbre de classification Standard (Synthétique)")
plt.tight_layout()
plt.savefig("arbre_standard_synth.png", dpi=300)  # Sauvegarder l'image en haute résolution
plt.show()

# Calcul des erreurs d'apprentissage et de test pour l'arbre standard
pred_train_standard = tree_standard.predict(X)
error_train_standard = 1 - accuracy_score(y, pred_train_standard)

X_test = data_test.drop(columns=['y'])
y_test = data_test['y']
pred_test_standard = tree_standard.predict(X_test)
error_test_standard = 1 - accuracy_score(y_test, pred_test_standard)

print(f"Erreur d'apprentissage (Standard) : {error_train_standard:.4f}")
print(f"Erreur de test (Standard) : {error_test_standard:.4f}")

# Matrice de confusion pour l'arbre standard
conf_matrix = confusion_matrix(y_test, pred_test_standard)
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Matrice de confusion - Arbre standard")
plt.colorbar()
tick_marks = np.arange(len(y.cat.categories))
plt.xticks(tick_marks, y.cat.categories, rotation=45)
plt.yticks(tick_marks, y.cat.categories)
plt.xlabel('Prédiction')
plt.ylabel('Vraie classe')
plt.tight_layout()
plt.savefig("confusion_matrix_standard.png", dpi=300)
plt.show()

# ---------------------------
# Arbre de classification avec profondeur maximale
# ---------------------------

# Arbre de décision avec profondeur maximale (par exemple, profondeur maximale de 5)
tree_max_depth = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_max_depth.fit(X, y)

# Visualiser l'arbre de décision avec profondeur maximale
plt.figure(figsize=(20, 10))
plot_tree(tree_max_depth, filled=True, feature_names=X.columns,
          class_names=[str(cls) for cls in y.cat.categories], rounded=True)
plt.title("Arbre de classification avec Profondeur Maximale (Synthétique)")
plt.tight_layout()
plt.savefig("arbre_max_depth_synth.png", dpi=300)
plt.show()

# Calcul des erreurs d'apprentissage et de test pour l'arbre avec profondeur maximale
pred_train_max_depth = tree_max_depth.predict(X)
error_train_max_depth = 1 - accuracy_score(y, pred_train_max_depth)

pred_test_max_depth = tree_max_depth.predict(X_test)
error_test_max_depth = 1 - accuracy_score(y_test, pred_test_max_depth)

print(f"Erreur d'apprentissage (Profondeur Maximale) : {error_train_max_depth:.4f}")
print(f"Erreur de test (Profondeur Maximale) : {error_test_max_depth:.4f}")

# Matrice de confusion pour l'arbre avec profondeur maximale
conf_matrix_md = confusion_matrix(y_test, pred_test_max_depth)
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix_md, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Matrice de confusion - Arbre avec profondeur maximale")
plt.colorbar()
plt.xticks(tick_marks, y.cat.categories, rotation=45)
plt.yticks(tick_marks, y.cat.categories)
plt.xlabel('Prédiction')
plt.ylabel('Vraie classe')
plt.tight_layout()
plt.savefig("confusion_matrix_max_depth.png", dpi=300)
plt.show()

# -------------------------------
# 2. Reconnaissance de chiffres manuscrits
# -------------------------------

try:
    # Charger les données ZIP
    zip_train = pd.read_csv("zip_train.txt", sep='\s+', header=None)
    zip_test = pd.read_csv("zip_test.txt", sep='\s+', header=None)

    # Nommer les colonnes
    zip_train.columns = ['y'] + [f'x{i+1}' for i in range(256)]
    zip_test.columns = ['y'] + [f'x{i+1}' for i in range(256)]
except FileNotFoundError as e:
    print(f"Erreur de chargement des fichiers ZIP: {e}")
    print("Vérifiez que les fichiers zip_train.txt et zip_test.txt sont présents dans le répertoire de travail.")
    exit(1)
except Exception as e:
    print(f"Erreur lors du chargement des données ZIP: {e}")
    exit(1)

# Afficher les premières lignes pour vérifier les données
print("Aperçu des données ZIP:")
print(zip_train.head())

# Transformer y en catégorie
zip_train['y'] = zip_train['y'].astype('category')
zip_test['y'] = zip_test['y'].astype('category')

# Vérification des valeurs manquantes
missing_count_zip_train = zip_train.isnull().sum().sum()
if missing_count_zip_train > 0:
    print(f"Attention: {missing_count_zip_train} valeurs manquantes détectées dans les données ZIP d'entraînement")

missing_count_zip_test = zip_test.isnull().sum().sum()
if missing_count_zip_test > 0:
    print(f"Attention: {missing_count_zip_test} valeurs manquantes détectées dans les données ZIP de test")

# Vérification des types de données des colonnes
print("Types de données des colonnes ZIP:")
print(zip_train.dtypes.value_counts())

# Fonction pour vérifier et convertir les colonnes d'images
def convert_to_numeric(data, image_columns):
    """
    Convertit les colonnes d'images en numérique, en gérant les erreurs.
    Retourne les données avec NaN pour les valeurs non numériques.
    """
    image_data = data[image_columns].copy()
    for col in image_columns:
        try:
            image_data[col] = pd.to_numeric(image_data[col], errors='coerce')
        except Exception as e:
            print(f"Erreur lors de la conversion de la colonne {col}: {e}")

    if image_data.isnull().any().any():
        nan_count = image_data.isnull().sum().sum()
        print(f"Avertissement: {nan_count} valeurs non numériques ont été trouvées dans les colonnes d'images et ont été remplacées par NaN.")
    return image_data

# Fonction pour s'assurer que les données d'images sont numériques
def clean_image_data(data):
    # Sélectionner uniquement les colonnes d'images (en ignorant la colonne 'y')
    image_columns = [col for col in data.columns if col != 'y']

    # Convertir les données d'images en numériques
    image_data = convert_to_numeric(data, image_columns)

    # Remplacer les valeurs manquantes éventuelles par la moyenne de la colonne
    for col in image_data.columns:
        if image_data[col].isnull().any():
            mean_val = image_data[col].mean()
            image_data[col].fillna(mean_val, inplace=True)
            print(f"Les valeurs manquantes dans la colonne {col} ont été remplacées par la moyenne ({mean_val:.2f}).")

    return image_data

# Visualiser les images pour chaque classe
def display_images(data, num_samples=1):
    """
    Affiche des exemples d'images pour chaque classe.

    Args:
        data: DataFrame contenant les données
        num_samples: Nombre d'images à afficher par classe
    """
    unique_classes = sorted(data['y'].unique())

    # Créer une grille pour afficher toutes les classes
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.flatten()

    for i, cls in enumerate(unique_classes):
        if i < len(axes):  # S'assurer qu'on ne dépasse pas le nombre d'axes
            # Filtrer les données pour la classe actuelle
            class_data = data[data['y'] == cls]

            # S'assurer qu'il y a suffisamment d'échantillons
            samples_to_show = min(num_samples, len(class_data))

            if samples_to_show > 0:
                # Sélectionner un échantillon
                sample_idx = np.random.choice(len(class_data), 1)[0]
                sample_image = class_data.iloc[sample_idx, 1:].values.astype(np.float64)

                # Mettre en forme comme une image 16x16
                image_matrix = sample_image.reshape(16, 16)

                # Affichage de l'image
                axes[i].imshow(image_matrix, cmap='gray', interpolation='nearest')
                axes[i].set_title(f"Classe {cls}")
                axes[i].axis('off')

    plt.tight_layout()
    plt.savefig("digit_samples.png", dpi=300)
    plt.show()

# Afficher un exemple de chaque chiffre
display_images(zip_train)

# Préparer les données pour l'entraînement de l'arbre de décision
X_zip = clean_image_data(zip_train)  # Nettoyer et convertir les données d'images
y_zip = zip_train['y']  # Labels

X_zip_test = clean_image_data(zip_test)
y_zip_test = zip_test['y']

# Créer l'arbre CART pour les chiffres manuscrits
# Utilisation d'un arbre régularisé pour éviter le surapprentissage
tree_zip = DecisionTreeClassifier(max_depth=15, min_samples_split=5, min_samples_leaf=2, random_state=42)
tree_zip.fit(X_zip, y_zip)

# Pour visualiser un arbre plus simple (pour la lisibilité), créons un modèle réduit
tree_zip_simple = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_zip_simple.fit(X_zip, y_zip)

# Visualiser l'arbre simplifié
plt.figure(figsize=(20, 10))
plot_tree(tree_zip_simple, filled=True, feature_names=X_zip.columns,
          class_names=[str(cls) for cls in y_zip.cat.categories], rounded=True)
plt.title("Arbre de classification simplifié pour les chiffres manuscrits")
plt.tight_layout()
plt.savefig("arbre_zip_simple.png", dpi=300)
plt.show()

# Prédictions et erreurs pour l'arbre complet
pred_zip_train = tree_zip.predict(X_zip)
pred_zip_test = tree_zip.predict(X_zip_test)

error_zip_train = 1 - accuracy_score(y_zip, pred_zip_train)
error_zip_test = 1 - accuracy_score(y_zip_test, pred_zip_test)

print(f"Erreur apprentissage (ZIP) : {error_zip_train:.4f}")
print(f"Erreur test (ZIP) : {error_zip_test:.4f}")

# Matrice de confusion pour les prédictions de test
conf_matrix_zip = confusion_matrix(y_zip_test, pred_zip_test)
plt.figure(figsize=(10, 8))
plt.imshow(conf_matrix_zip, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Matrice de confusion - Reconnaissance de chiffres")
plt.colorbar()
tick_marks = np.arange(len(y_zip_test.cat.categories))
plt.xticks(tick_marks, y_zip_test.cat.categories, rotation=45)
plt.yticks(tick_marks, y_zip_test.cat.categories)
plt.xlabel('Prédiction')
plt.ylabel('Vraie classe')
plt.tight_layout()
plt.savefig("confusion_matrix_zip.png", dpi=300)
plt.show()

# Erreur par chiffre (test)
print("\nErreur par chiffre (test):")
for chiffre in sorted(zip_test['y'].unique()):
    idx = zip_test['y'] == chiffre
    err = 1 - accuracy_score(zip_test.loc[idx, 'y'], pred_zip_test[idx])
    count = sum(idx)
    print(f"Chiffre {chiffre}: erreur = {err:.4f} (sur {count} exemples)")

# Identifier les chiffres les plus confondus
print("\nPaires de chiffres les plus confondues:")
n_classes = len(y_zip_test.cat.categories)
confusion_pairs = []

for i in range(n_classes):
    for j in range(n_classes):
        if i != j:  # Éviter la diagonale
            confusion_pairs.append((i, j, conf_matrix_zip[i, j]))

# Trier les paires par nombre de confusions (décroissant)
confusion_pairs.sort(key=lambda x: x[2], reverse=True)

# Afficher les 5 paires les plus confondues
for i, j, count in confusion_pairs[:5]:
    print(f"Chiffre {i} prédit comme {j}: {count} fois")
```
